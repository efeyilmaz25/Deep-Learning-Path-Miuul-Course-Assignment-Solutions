````python
from deepface import DeepFace
import os
import pandas as pd
import cv2
import matplotlib.pyplot as plt
````

````python
def analyze_images_in_directory(root_path):
  image_paths = [os.path.join(root_path, file) for file in os.listdir(root_path) if file.endswith(('.jpg', '.jpeg', '.png'))]
  results = []

  for image_path in image_paths:
    result = DeepFace.analyze(img_path=image_path, actions=['age', 'gender', 'race', 'emotion'])

    for idx, face in enumerate(result):
      results.append({
          'file_name': os.path.basename(image_path),
          'age': face['age'],
          'gender': face['dominant_gender'],
          'race': face['dominant_race'],
          'emotion': face['dominant_emotion']
      })

  return results
````

We have written our function. Now let's provide the path information that we will work on and store our results in the 'analyze_results' variable. Finally, let's print these results with a simple for loop.

````python
path = "/content/face_assignment"
````

````python
analyze_results = analyze_images_in_directory(path)
````

````python
for result in analyze_results:
  print(result)
````

# **Face Recognition**

Face Recognition is a technology used in artificial intelligence and computer vision. It analyzes people's facial features to identify or verify them. This technology is often used for security, biometric verification, and user experience enhancement.

**HOW DOES FACE RECOGNİTİON WORK?**

**1-) Face Detection:** The system detects human faces in an image or video and extracts certain points of the face (such as eyes, nose, mouth). In this process, deep learning techniques such as Convolutional Neural Networks (CNN) are often used.

**2-) Face Recognition:** The detected face is compared with previously recorded faces. At this stage, algorithms that extract facial features (e.g. OpenFace, FaceNet, Dlib) are used. To describe a face, the algorithm creates vectors using the geometric structure of the face, specific key points, and the distances between those points. This is called face embedding.

**3-) Comparison and Authentication:** Authentication is performed by comparing the created face vector with other face vectors in the database.

**METHODS USED IN FACE RECOGNİTİON TECHNOLOGIES**

* **Deep Learning Based Methods:** The most commonly used face recognition methods are based on deep learning. Especially CNN-based models such as FaceNet and ArcFace provide high accuracy.

* **Haar Cascades:** An old method used in face detection. It is widely used in object detection tasks, but is not as accurate as modern deep learning methods.

* **LBPH (Local Binary Patterns Histograms):** It is one of the earlier face recognition techniques, but it is not as powerful and flexible as deep learning models.





# **DeepFace**

DeepFace is a popular Python library used in face recognition and face verification, and is widely used especially in social media platforms and security applications. This deep learning-based library was developed by **Facebook** and stands out with its high accuracy rates. It achieves results very close to human face recognition ability with an **accuracy rate of 97.35%.** DeepFace offers a wide range of users by supporting different face recognition models such as **VGG-Face**,**Google** **FaceNet**, **OpenFace**, **DeepID**, **ArcFace** and **Dlib**. Working with these models, the library can perform operations such as face recognition, face verification, gender, age and emotion analysis. In addition, it detects faces in images using powerful face detection algorithms such as **MTCNN (Multi-task Cascaded Convolutional Networks)** and aligns these faces for analysis. It extracts face embeddings with deep learning-based models and calculates the similarity of faces using these features.
